```python
data["STA"]= pd.to_datetime(data['STA'])
data['AM/PM']=data['STA'].dt.strftime('%Y-%m-%d %I:%M:%S %p')
data["STD"]= pd.to_datetime(data['STD'])
data['STD']=data['STD'].dt.strftime('%m')
data.rename(columns={'STD': 'Month'}, inplace=True)


data.rename(columns={'STA': 'Seasons'}, inplace=True)
data['Seasons']=data['Month'].apply(lambda x: 'Alto'if x in ['01','04','07','08','12']else 'Normal')

data.head()

```

```python
import statsmodels.api as sm
# Preparamos los datos para el modelo
X = data['Bookings']  # Variable independiente
y = data['Passengers']  # Variable dependiente

# Añadimos una constante al modelo, ya que statsmodels no lo hace automáticamente
X = sm.add_constant(X)

# Creamos el modelo de regresión lineal y lo ajustamos a los datos
modelo = sm.OLS(y, X).fit()

# Mostramos el resumen del modelo
print(modelo.summary())

```
